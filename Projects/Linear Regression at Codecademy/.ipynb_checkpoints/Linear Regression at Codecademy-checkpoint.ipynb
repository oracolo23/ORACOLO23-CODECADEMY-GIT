# Linear Regression at Codecademy


```python
# Load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats as stats
import statsmodels.api as sm


# Read in the data
codecademy = pd.read_csv('codecademy.csv')
print(codecademy.head(10))
```

## REGRESSION MODEL - QUANTITATIVE PREDICTOR



###  1. CHECKING FOR POSSIBILE LINEAR RELATIONSHIP


```python
plt.scatter(codecademy.completed, codecademy.score)
plt.title('Scatterplot of Completed Items VS Score')
plt.ylabel('Score')
plt.xlabel('Completed Items')
plt.show()
```

### 2. FIT LINEAR MODEL, PREDICT SCORE BASED ON COMPLETED ITEMS


```python
# using sm.OLS.from_formula to build a model with age as predictor

model = sm.OLS.from_formula('score~completed', codecademy)
results = model.fit()

# printing parameters and prediting score for 60 completed items to find best line for linear regression
print(results.params)
y_inter = results.params[0]
slope = results.params[1]

pred_score60 = slope * 60 + y_inter
print(pred_score60)

```

### 3. PLOT REGRESSION LINE USING COEFFICENTS FROM LINEAR MODEL


```python
#plotting data to visualize linear regression line on top of the scatterplot
plt.scatter(codecademy.completed, codecademy.score)
plt.title('Scatterplot of Completed Items VS Score')
plt.plot([0, 60],[results.params[0], pred_score60 ], color ='r')
plt.ylabel('Score')
plt.xlabel('Completed Items')
plt.show()
```

As we can see from above output, we can observ the following : 

- y_intercept value is 13,21 we can interpret this saying that a students who complete 0 items is predicted to score 13,2

- the slope is 1,30 , so for every units (completed items) there is an increment of 1,3 in score

- There is linear association between those variables

We are now going to verify Normality and Homoscedasticity assumptions calculating fitted values and residuals.

### 4. CALCULATING FITTED VALUES AND RESIDUALS AND PLOT THEM TO CHECK IF NORMALITY AND HOMOSCEDASTICITY ASSUMPTIONS ARE SATISFIED


```python
fitted= results.predict(codecademy)
residuals = codecademy.score - fitted

plt.title('Histogram of residuals to check Normality Assumption')

plt.hist(residuals)
plt.axvline(residuals.mean(), color='r', linestyle='dashed')
plt.axvline(residuals.median(), color='y', linestyle='dashed')
print('That residuals mean is : ' + str(residuals.mean()))
print('That is equal to : ' + str(format(residuals.mean(), '.16f')))

print('That residuals median is : ' + str(residuals.median()))


plt.show()
plt.clf()

plt.scatter(x=fitted, y=residuals)
plt.title('Scatterplot of Fitted Values VS Residuals')
plt.axhline(y=0, color='r', linestyle='dashed')

plt.ylabel('Residuals')
plt.xlabel('Fitted')
plt.show()
```

##  REGRESSION MODEL - BINARY CATEGORICAL PREDICTOR


```python
# Building a model to predict time spent based on lesson

model = sm.OLS.from_formula('score~lesson', codecademy)
results = model.fit()
print(results.params)

# Calculating average score mean for each lesson

codeA= codecademy[codecademy['lesson']=='Lesson A']
codeB= codecademy[codecademy['lesson']=='Lesson B']

print('Lesson A Score Mean is: ' + str(codeA.score.mean()))
print('Lesson B Score Mean is: ' + str(codeB.score.mean()))
```

```python
#Plotting data to show datapoints and linear regression

plt.scatter(codecademy.lesson, codecademy.score)
plt.plot([0, 1], [codeA.score.mean(), codeB.score.mean()], color='r')
plt.show()
```
