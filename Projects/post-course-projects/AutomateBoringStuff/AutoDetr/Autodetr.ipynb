```python
import pandas as pd
import csv
```

```python
# Importing data from Excel file,  Sheet name 'Virgin'
data = pd.read_excel("/Users/matteoamanti/Desktop/Macr.xlsm", sheet_name=['Virgin'])
```

```python
data = data["Virgin"]
```

```python
#Checking data
data.head(10)
```

```python
# Importing IDs of rows we need to remove 
exceptions = pd.read_excel("/Users/matteoamanti/Desktop/Macr.xlsm", sheet_name=['EXCEPTIONS'])
exceptions = exceptions['EXCEPTIONS']

credit_requests = pd.read_excel("/Users/matteoamanti/Desktop/Macr.xlsm", sheet_name=['CREDIT REQUESTS'])
requests = credit_requests['CREDIT REQUESTS']
```

```python
# Setting ID column as index 
exceptions = exceptions.set_index('ID')
exceptions.head()
```

```python
# Setting ID column as index 
requests = requests.set_index('ID')
requests.head()
```

```python
# Merging DataFrames of IDs to remove an printing its lenght
to_remove = pd.concat([exceptions,requests])
len(to_remove)
```

```python
# Applying desired filters to extract PVs with negative balance
filtered_data = data.loc[(data.TYPE=="PV") & (data.STATUS=="ATTIVO") & (data.GB == 0)]
```

```python
# Sorting by 
filtered_data= filtered_data.sort_values('saldo dovuto', ascending = False).set_index('ID')
filtered_data.head()
```

```python
to_remove.index
```

```python
print('Current length: ' + str(len(filtered_data)))
for exceptions in to_remove.index:
   
    try:
        filtered_data = filtered_data.drop(exceptions)
        print('ID {} has been successfully removed from list'.format(exceptions))
    except KeyError:
        print('ID {} is not in list'.format(exceptions))
print('Current length: ' + str(len(filtered_data)))
```

```python
filtered_data = filtered_data.reset_index().rename(columns={'index':'ID'}).sort_values('saldo progressivo', ascending = False)
filtered_data.head()
```

```python
with open('/Users/matteoamanti/Desktop/EmailComposer/AutoDetr/cleaned_list.csv', 'w') as output:
    writer = csv.writer(output)
    header=filtered_data.columns
    data = filtered_data.values
    # write the header
    writer.writerow(header)

    # write the data
    writer.writerows(data)
    
```
