```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
import numpy as np
```

```python
links_dict = {}
def dict_builder(list_of_links):
    i=0
    for links in list_of_links:
        index = i
        item = list_of_links[i]
        link = item.get('href')
        title= item.get('title')
        dict_item = {}
        dict_item['title'] = title
        dict_item['link'] = link
        links_dict[i] = dict_item
        i+=1
    return links_dict
```

```python
complete_list = []
def html_retreiver(city,n_pages):
    # the function will accept int representing the n of desired pages to retreive and the target city
    
    
    for i in range(n_pages):
        html = requests.get('https://www.immobiliare.it/vendita-case/{}/?criterio=prezzo&ordine=desc&pag={}'.format(city,i))
        contents = html.content
        soup = BeautifulSoup(contents, 'html.parser')
        home_lists = soup.findAll('a',class_='in-card__title', href=True)
        for item in home_lists:
            complete_list.append(item)
        
        i+=1
```

```python
html_retreiver('roma',200)
```

```python
dict_builder(complete_list)
```

```python
def extract_info(link):
    
    html_link = requests.get(link)
    contents_link = html_link.content
    soup_link = BeautifulSoup(contents_link, 'html.parser')
    
    items_title = soup_link.findAll('h1', attrs={'class':['in-titleBlock__title']})  
    items_loc = soup_link.findAll('span', attrs={'class':['in-location']})  
    
    
    try:
        item_dict ={}
        item_dict['city'] = items_loc[0].get_text()
        item_dict['borough'] = items_loc[1].get_text()
        item_dict['title'] = items_title[0].get_text()

    except IndexError:
        print('Errore retreving loc info')
    finally:
        
        item_dict['link'] = link
        titles = soup_link.findAll('dt')
        data = soup_link.findAll('dd')
        i=0
        item={}
        for title in titles:
            item_dict[titles[i].get_text()] = data[i].get_text()
            i+=1
    return item_dict
    
    
    
```

```python
nan_proportions = ((df2.count() / len(df2))*100).sort_values(ascending=False)
to_drop = nan_proportions[nan_proportions < 10].index.to_list()
df2=df2.drop(labels=to_drop, axis = 1).fillna(0)
df2=df2.drop(df2[df2.prezzo == 0].index).reset_index(drop=True)
df2.prezzo = df2.prezzo.str.replace('€','')
df2.prezzo = df2.prezzo.str.replace('.','').astype(int)
```

```python
df2['Riferimento e Data annuncio'] = df2['Riferimento e Data annuncio'].str.slice(start=-10)
```

```python
df2 = df2.rename(columns={'Riferimento e Data annuncio':'Data Annuncio'})
```

```python
df2['totale piani edificio'] = df2['totale piani edificio'].str.slice(stop=2).str.strip()
```

```python
columns = df2.columns.to_list()
new_list = []
for item in columns:
    item = item.title().replace(' ','')
    new_list.append(item)
```

```python
df2.columns = new_list
df2.Tipologia = df2.Tipologia.apply(lambda x: x.split(' | '))
df2 = df2.rename(columns={'Tipologia':'toRemove'})
```

```python
# initialize columns
df2['Tipologia'] = np.nan
df2['Proprietà'] = np.nan
df2['ClasseImmobiliare'] = np.nan
df2.head()
```

```python
props = ['Intera proprietà','Nuda proprietà']
```

```python


i = 0
for features in df2.toRemove:
    type_ = features[0]
    
    if len(features) == 3:
        prop = features[1]
        class_ = features[2]
    elif len(features) ==1:
        prop = np.nan
        class_ = np.nan
    else:
        if features[1] in props:
            prop = features[1]
            class_ = np.nan
            
        else :
            prop = np.nan
            class_ = features[1]
   
    df2.loc[i, 'Tipologia']= type_
    df2.loc[i, 'Proprietà']= prop
    df2.loc[i, 'ClasseImmobiliare']=  class_    
    i+=1
```

```python
def has_feature(feature, strings):
    
    if hasattr(strings, '__iter__'): 
        return feature in strings
    else:
        return False

# Testing function
has_feature('ascensore',df2.loc[2].Piano)
```
