```python
# loading libraries and data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats

data = pd.read_csv('all_data.csv')
#Checking shape of df
print(data.shape)
print(data.head())
```

```python
# Checking unique values for Country ad Years
years = data.Year.unique()
countries = data.Country.unique()
#Replacing long name in country column
data['Country'] = data['Country'].str.replace('United States of America', 'USA')
print(data.info())
print(data.drop('Year', axis=1).describe())
```

```python
# Histogram of GDP values distribution
plt.figure(figsize=(12,6))
plt.hist(data.GDP, bins=30)
plt.axvline(data.GDP.median(), color = 'r', label='Median')
plt.axvline(data.GDP.mean(),color='y', label='Average')
plt.title('Yearly GDP Distrbution in Trillions ($)', size=20, pad=20)
plt.legend()
plt.show()
```

# Univariate Analysys

```python

# Creating aggreagated df with mean values
data_mean = data.drop('Year', axis=1).groupby('Country', as_index=False).mean()

# Barplot of average LEABY
sns.catplot(x="Life expectancy at birth (years)", y='Country',
                data=data_mean,
                kind="bar", 
                aspect=3,
                height=5);
plt.title('Average Life expectancy at birth [2000/2015]', size=20)
plt.show()
```

```python

# LEABY lineplot
sns.relplot(x='Year', y='Life expectancy at birth (years)',
            data=data,
            hue='Country',
            kind='line',
            aspect=3)
plt.title('Life expectancy at birth over timeframe 2000/2015', size=20)
plt.show()
```

```python

# GDP Histplot by Year , Hue == country
sns.catplot(x="Year", y="GDP",
                hue="Country",
                hue_order=data.groupby('Country').sum('GDP').sort_values('GDP').index, 
                data=data, kind="bar", 
                aspect=3,
                height=5);
plt.title('Gross Domestic Product by Year [2000/2015]', size=20)
plt.show()
```

```python
# GDP Histplot by Country , Hue == Year

sns.catplot(x="Country", y="GDP",
                hue='Year',
                data=data, kind="bar",
                # sorting values for yearly aggregated GDP
                order=data.groupby('Country').sum('GDP').sort_values('GDP').index,
                aspect=3,
                height=5);
plt.title('Gross Domestic Product by Country [2000/2015]', size=20)
plt.show()
```

```python
graphGDP = sns.FacetGrid(data, col="Country", col_wrap=3, sharey = False)

graphGDP = (graphGDP.map(sns.lineplot,"Year","GDP")
         .add_legend()
         .set_axis_labels("Year","GDP in Trillions of U.S. Dollars"))

graphGDP.fig.suptitle('GDP for each country over years', y=1.05, size=20)
plt.subplots_adjust(hspace=0.4, wspace=0.4)
plt.show()
#graphGDP;

graphLEABY = sns.FacetGrid(data, col="Country", col_wrap=3, sharey = False)

graphLEABY = (graphLEABY.map(sns.lineplot,"Year","Life expectancy at birth (years)")
         .add_legend()
         .set_axis_labels("Year","Life expectancy at birth (years)"))
graphLEABY.fig.suptitle('LEABY for each country over years',y=1.05, size=20)

plt.show()
#graphLEABY;
```

# Bivariate analysis and correlation check

```python
# Scatterplot + Lineplot of GDP vs Life Exp  for each country over years ('00/'15) to check for patterns
for country in countries:
    datatemp = data[data.Country == country]
    r, p = stats.pearsonr(datatemp['GDP'], datatemp['Life expectancy at birth (years)'])
    
    # Scatterplot
    sns.relplot(x='GDP', y='Life expectancy at birth (years)',
                data=datatemp,
                hue='Year',
                size='GDP',
                kind='scatter',
                aspect=1   )
    plt.title('{}'.format(country), size=20)
    plt.show()
    # Printing Pearsonr coefficent and related p-value for each Series per country (GDP ~ LEABY)
    print('Pearson Correlation Coefficent for {} is :'.format(country.upper()) + str(round(r,3)),'p-value: ' + str(p))

    # Lineplot with shared X axis,  GDP on y1, LEABY on y2
    fig, ax_left = plt.subplots(figsize=(10,4))
    ax_right = ax_left.twinx()
    ax_left.plot(datatemp.Year, datatemp['GDP'], color='red', label='GDP',marker='*')
    ax_left.set_ylabel('GDP')
    ax_left.legend()
    ax_right.plot(datatemp.Year, datatemp['Life expectancy at birth (years)'], color='orange', label='Life Exp.', marker='o')
    ax_right.set_ylabel('Life Exp')
    ax_right.legend(loc=4)
    plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=data['Life expectancy at birth (years)'],
                y=data.GDP,
                hue=data.Country).legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)

plt.title('General Scatterplot of GDP vs LEABY ', size=20, pad=20)
plt.show()
```

Generally speaking, we have observed the following : 


- GDP increase is strongly associated (0.90 or more) with LEABY in our dataset
- Each nation shows increasing values for both GDP and LEABY over the observed  timeframe 2000~2015 
- Downside during 2008 financial crisis for GDP in every observed state

We also observed that Zimbawe had the most LEABY increase during the given timeframe, going from 50 to 60.
