```python

# Import necessary libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
# load rankings data
wood = pd.read_csv('Golden_Ticket_Award_Winners_Wood.csv')
# load rankings data
steel = pd.read_csv('Golden_Ticket_Award_Winners_Steel.csv')

wood
```

## 1. Function that will plot the ranking of a given roller coaster over time as a line. 

   

```python
# Create a function to plot rankings over time for 1 roller coaster

def rank_plotter(name,location,df):
    # Filtering rows matching Name and Location
    rows = df.loc[(df.Name == name) & (wood.Location == location)]
    # Plotting rank as line
    plt.figure(figsize=(10,5))
    plt.plot(rows['Year of Rank'], rows.Rank)
    # Setting Y and X ticks to get values from filtered rows ( max/min rank  and  years if availble)
    plt.yticks(range((min(rows.Rank)-1),max(rows.Rank)+1), size = 14)   
    plt.xticks(rows['Year of Rank'],rotation=30, size = 14)
    # Title & Axes labels
    plt.title('Roller Coaster Name: {}, \n Location: {}'.format(name,location), size=20)
    plt.xlabel('Year', size = 12)
    plt.ylabel('Rank',size = 12)
    plt.show()
    
# Create a plot of El Toro ranking over time
rank_plotter('El Toro', 'Jackson, N.J.', wood)
```

##  2. Function that will plot the ranking of 2 given roller coaster 

```python
# Create a function to plot rankings over time for 2 specific roller coaster

def rank_plotter_2(name1,location1,name2,location2,df):
    
    # Filtering rows matching Name and Location for 1
    rows1 = df.loc[(df.Name == name1) & (wood.Location == location1)]
    rows2 = df.loc[(df.Name == name2) & (wood.Location == location2)]
    
    # Plotting 2 ranks as line on the same figure
    plt.figure(figsize=(10,5))
    plt.plot(rows1['Year of Rank'], rows1.Rank,label = name1)
    plt.plot(rows2['Year of Rank'], rows2.Rank,label = name2)
    
    # Setting Y ticks to check for minimum and maximum value for both ranks
    plt.yticks(range((min(min(rows1.Rank),min(rows2.Rank))-1),max(max(rows1.Rank),max(rows2.Rank))+1), size = 14)  
    
    # Title & Axes labels
    plt.title('Roller Coaster 1 Name: {}, \n Location: {}, \nRoller Coaster 2 Name: {}, \n Location: {}'.format(name1,location1,name2,location2), size=20)
    plt.xlabel('Year', size = 12)
    plt.ylabel('Rank',size = 10)
    plt.legend()
    plt.show()
    
rank_plotter_2('El Toro', 'Jackson, N.J.','The Voyage','Santa Claus, Ind.',wood)
rank_plotter_2('Phoenix', 'Elysburg, Pa.','Shivering Timbers','Muskegon, Mich.',wood)
```

## 3. Function that will plot the ranking of N roller coaster,  N will be provided as parameter

```python
sns.set_theme(style="darkgrid")

def top_n_plotter(n, df):
        top_n= df[df.Rank<=n]
        fig = plt.figure(figsize=(15,10))
        plt.yticks(range(1,n+1))
        ax = sns.lineplot(x=top_n['Year of Rank'], y = top_n.Rank, hue='Name',data=top_n,palette="bright")
        sns.move_legend(ax, "upper right",bbox_to_anchor=(1.2, 1))
        plt.title(' TOP {} ROLLER COASTER RANK ~ LAST 5 YEARS '.format(n), size=20)
        plt.show()
top_n_plotter(15, wood)
```

# Roller Coaster data analysis

[Captain Coaster](https://captaincoaster.com/en/) Data on all roller coasters documented on Captain Coaster has been accessed through its API and stored in `roller_coasters.csv`. 

```python
roller_coaster = pd.read_csv('roller_coasters.csv')
roller_coaster
```

## 1. Features Data Analisys

```python
# Create a function to plot histogram of column values

def feature_histogram(dataframe, column):
    feature = dataframe[column].dropna() 
    plt.figure(figsize=(10,5))
    plt.hist(feature)
    plt.title('{} Histogram'.format(column,),size = 20)
    plt.show()
    
    
# Histogram of roller coaster speed   
feature_histogram(roller_coaster, 'speed') 
```

```python
# Histogram of roller coaster length
feature_histogram(roller_coaster, 'length')
```

```python
# Histogram of roller coaster number of inversions
feature_histogram(roller_coaster, 'num_inversions')
```

```python
# height values have some outliers that we need to handle, index 683 , 730  & 2166. We will drop these values when plotting our histogram
print(roller_coaster.height.nlargest(10))
plt.hist(roller_coaster.height.dropna())
plt.show()
```

```python
heights = roller_coaster.height.drop([683,730,2166]).dropna()
plt.figure(figsize=(10,5))
plt.hist(heights)
plt.title(' Height Histogram', size = 20)
plt.show()
```

## 2. num_inversion  by Park

```python
#Grouping coasters by park in order to extract top5 parks by coasters count

grouped = roller_coaster.groupby('park').count()
grouped_top5 = grouped.sort_values('name', ascending = False).nlargest(6, columns='name')

# Dropping Foire coaster as we want to check parks only
grouped_top5 = grouped_top5.drop('Foire')


# Create a function to plot inversions by coaster at park
import seaborn as sns

def feature_barplot(dataframe,park_name, column):
    data = dataframe[dataframe.park == park_name]
    plt.figure(figsize=(25,5))
    plt.title('[{}] Coasters \n{} Barplot'.format(park_name,column,),size = 20)
    sns.barplot(data=data, x='name', y=column)
    plt.xticks(rotation=30)
    plt.show()




# Storing indices as list 
park_list = grouped_top5.index.tolist()
grouped_top5
```

```python
# Create barplot of inversions by roller coasters

for parks in park_list:
    feature_barplot(dataframe = roller_coaster,park_name = parks, column='num_inversions')
```

## 3. Operating Status Overview 

```python

# Create a function to plot a pie chart of status.operating

# Grouping by status
proportions = roller_coaster[['name','status']].groupby('status',as_index=False).count()
# Normalize proportions
proportions.name = round(proportions.name/len(roller_coaster),2)
proportions = proportions.rename(columns={'name':'proportion'})
# Setting mask 'Other' for lables under 5%
proportions.status=proportions.status.mask(proportions.proportion < .05, other='Other')
proportions = proportions.groupby('status').sum().sort_values('proportion', ascending=False)
proportions


# Create pie chart of roller coasters
plt.figure(figsize=(10,10))
plt.pie(proportions.proportion, labels = proportions.index, normalize = True, autopct='%.2f%%',textprops={'fontsize': 20})
plt.axis('Equal')
plt.title('Coasters Status Overview', size=25)
plt.tight_layout()
plt.show()
```

## 4. Features relationship

```python
# Dropping outliers row
roller_coster_no_ol = roller_coaster.drop([683,730,])


# Create a function to plot scatter of 2 features in our dataframe
def features_scatterplot(dataframe, feature1, feature2):
    
    plt.figure(figsize=(15,10))
    sns.regplot(data = dataframe, x = dataframe[feature1], y = dataframe[feature2], line_kws={'color':'r', 'label':'regression line'})
    plt.xlabel(feature1, size =15)
    plt.ylabel(feature2, size =15)
    plt.legend()
    
# Create a scatter plot of roller coaster height by speed
features_scatterplot(roller_coster_no_ol,'height','speed')
```

```python
features_scatterplot(roller_coster_no_ol,'length','speed')
```

```python
# Checking findings in correlation matrix for our dataframe ( outliers excluded )
roller_coster_no_ol.corr()
```

## Findings and final observations

Less than 80% of coasters in our dataframe are currently operating.

Features values for height, lenght and speed are normally distributed with values centered at the mean.

num_inversion instead shows a strong left skeweness with most of the values = 0

'height' Vs. 'speed' Scatterplot shows clear relationship between the two variables, higher speeds are associated with higher heights

'length' Vs. 'speed' Scatterplot shows weaker correlation the two variables, higher lenghts are often associated with higher heights

No further significant correlation is observed.
