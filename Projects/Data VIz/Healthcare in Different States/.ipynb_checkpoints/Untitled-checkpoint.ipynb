```python
#Lodading libraries and importing data
import numpy as np
import pandas as pd 
import seaborn as sns
from matplotlib import pyplot as plt

data = pd.read_csv('data.csv')
data.head()
```

Weâ€™re going to focus on only the ways in which hospitals charge patients for **Chest pain**.

We will print healthcare["DRG Definition"].unique() to see all of the different diagnoses in our dataset.

```python
#Extracting and checking procedures names from related column
procedures = data['DRG Definition'].unique()
procedures[:10]
```

```python
#Generating dataframe with rows containing '313 - CHEST PAIN' as procedure and then extracting only AL (Alabama) hospitals

chest_pain = data[data['DRG Definition'] == '313 - CHEST PAIN']
alabama_chest_pain = chest_pain[chest_pain['Provider State'] == "AL"]
```

```python
alabama_chest_pain_costs = alabama_chest_pain[' Average Covered Charges '].values
alabama_chest_pain_costs
```

```python
plt.boxplot(alabama_chest_pain_costs)
plt.show()
```

```python
#Repeating previous steps in a loop to extract all data for each state and creating 50 different boxplot

states = data['Provider State'].unique()
all_costs = []
for state in states : 
    state_chest_pain = chest_pain[chest_pain['Provider State'] == state]
    chest_pain_costs = state_chest_pain[' Average Covered Charges '].values
    all_costs.append(chest_pain_costs)

plt.figure(figsize=(20,8))    
plt.boxplot(all_costs, labels=states)
plt.show()
```

```python
#Ordering rows by Average Covered Charges values to plot most expensive states and regions

chest_pain = chest_pain.sort_values('Hospital Referral Region Description')

#Cleaning Hospital Rererral Regin Description column to remove Provider State tag

chest_pain['Hospital Referral Region Description']=chest_pain['Hospital Referral Region Description'].str.slice(start=4)
```

```python
chest_pain
```

```python
# Removing white space from columns name
states = data['Provider State'].unique()
columns = chest_pain.columns.str.strip()
chest_pain.columns=columns
chest_pain.columns
```

```python
# Dropping undesired columns
try:
    chest_pain = chest_pain.drop(columns=['DRG Definition', 'Provider Id', 'Provider Name','Provider Street Address','Provider City','Provider Zip Code','Total Discharges', 'Average Total Payments','Average Medicare Payments'], axis=1)
except:
    pass
```

```python
#Grouping by state and ordering to check top 15 most expensive states

states_data = chest_pain.groupby('Provider State',as_index=False).mean('Average Covered Charges').sort_values('Average Covered Charges', ascending=False).reset_index(drop=True)
top_15_states = states_data.nlargest(15, columns='Average Covered Charges') 
```

```python
top_15_states
```

```python
plt.figure(figsize=(25,8))
sns.barplot(data=top_15_states, y='Average Covered Charges', x='Provider State')
plt.title('MOST EXPENSIVE STATES FOR CHEST PAIN CHARGES', size=20)
plt.show()
```

```python
#Repeating same steps for Regions
regions_data = chest_pain.groupby('Hospital Referral Region Description',as_index=False).mean('Average Covered Charges').sort_values('Average Covered Charges', ascending=False).reset_index(drop=True)
top_30_region = regions_data.nlargest(30, columns='Average Covered Charges') 
top_30_region
```

```python
plt.figure(figsize=(40,8))
sns.barplot(data=top_30_region, y='Average Covered Charges', x='Hospital Referral Region Description')
plt.title('MOST EXPENSIVE REGIONS FOR CHEST PAIN CHARGES', size=20)
plt.xticks(rotation = 45)
plt.show()
```
