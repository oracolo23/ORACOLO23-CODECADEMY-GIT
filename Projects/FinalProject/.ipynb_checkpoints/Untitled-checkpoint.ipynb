# Credit/Debit  Cards Adoption Across the World

In this analysis we are going to explore potential relationship between population, genre, location and credit/debit card adoption.


Data has been gathered from  https://www.worldbank.org/en/home  

## 1. Loading And Cleaning Data

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python

#Importing and loading data into a Pandas dataset, replacing '..' datapoints with NaN values in order to drop them
data = pd.read_csv('/Users/matteoamanti/Desktop/Github/ORACOLO23-CODECADEMY-GIT/Projects/FinalProject/all_data.csv', delimiter = ',')
data = data.replace('..', np.nan).dropna()
data.head(12)
```

```python
data = data.drop(columns=['Series Code']).sort_values('Country Name')
data
```

```python
#Renaming columns
data.columns = ['Series Name', 'Country Name', 'Country Code', '2014','2017']
data.head(20)
```

```python
# Let's split dataframes in two different ones based on year

data_2014 = data.drop(columns='2017')
data_2017 = data.drop(columns='2014')
```

## 2. Trasforming Data

```python
# We will create a pivot table to trasform all 'Series Name' rows into columns for each year dataset 

data_2014_piv = data_2014.drop(columns='Country Code').pivot(index = 'Country Name', columns='Series Name').dropna()
data_2014_piv.columns = data_2014_piv.columns.droplevel(0)
data_2014_piv
```

```python
data_2017_piv= data_2017.drop(columns='Country Code').pivot(index = 'Country Name', columns='Series Name').dropna()
data_2017_piv.columns = data_2017_piv.columns.droplevel(0)
data_2017_piv
```

```python
data_2014_piv.columns
```

```python
# rename columns with shorter ones

new_columns = ['c_card_own','c_card_own_f','c_card_own_m','d_card_own','d_card_own_f','d_card_own_m','pop_f','pop_m','pop_tot', 'rur_pop', 'urb_pop']
data_2014_piv.columns =new_columns
data_2017_piv.columns =new_columns
```

```python
# Creating two lists splitting columns names for the values that need to be rounded

columns_no_round = ['pop_f','pop_m','pop_tot']
columns_round = data_2014_piv.columns.to_list() 

for items in columns_no_round:
            columns_round.remove(items)
    

```

```python
# Casting df with correct dtypes for both years


# 2014
data_2014_piv = data_2014_piv.astype(float)   
data_2014_piv[columns_no_round] = data_2014_piv[columns_no_round].astype(int) 
data_2014_piv.info()
# Population data will be dispplayed in Millions 
data_2014_piv[columns_no_round]  = data_2014_piv[columns_no_round].apply(lambda x: x/1000000)
data_2014_piv = data_2014_piv.round(2)


# 2017
data_2017_piv = data_2017_piv.astype(float)   
data_2017_piv[columns_no_round] = data_2017_piv[columns_no_round].astype(int) 
data_2017_piv.info()
data_2017_piv[columns_no_round]  = data_2017_piv[columns_no_round].apply(lambda x: x/1000000)
data_2017_piv = data_2017_piv.round(2)
```

```python
world_countries = pd.read_csv('/Users/matteoamanti/Desktop/Github/ORACOLO23-CODECADEMY-GIT/Projects/FinalProject/countries of the world.csv')
```

```python
world_countries
```

```python
# Extracting desired columns to merge with current df
regions = world_countries[['Country', 'Region']]
# Checking dtypes
regions.info()
```

```python
# We need to use strip() func and remove extra white space to make the merge work
regions = regions.apply(lambda x: x.str.strip())
```

```python
# Creating new index column and renaming the column to merge 
data_2014_piv = data_2014_piv.reset_index().rename(columns={'Country Name': 'Country'})
```

```python
data_2014_piv
```

```python
data_2014_piv_merged= data_2014_piv.merge(regions, how='left', on='Country').dropna()
data_2014_piv_merged
```

```python
# Calulating average cards ownership and generating new 'avg_card_own' column
data_2014_piv_merged['avg_card_own'] = data_2014_piv_merged[['c_card_own', 'd_card_own']].mean(axis=1)

# Moving new column columns to index 1
col_move = data_2014_piv_merged.pop('avg_card_own')
data_2014_piv_merged.insert(1,'avg_card_own', col_move)
# Moving Region column columns to index 1
col_move = data_2014_piv_merged.pop('Region')
data_2014_piv_merged.insert(1,'Region', col_move)
data_2014_piv_merged
```

```python
#Grouping for Macro-Region

grouped = data_2014_piv_merged.groupby('Region').mean().sort_values('avg_card_own', ascending = False)
grouped
```

```python
plt.figure(figsize=(20,10))
plt.bar(data = grouped, x = grouped.index, height = 'avg_card_own')
plt.axhline(grouped.avg_card_own.mean(), color = 'r')
plt.xticks(rotation=30)
plt.title('Average Credit/Debit Card Ownership Across The World', size = 20)
plt.show()
```

```python
data_2014_piv_merged
```

```python

fig, axs = plt.subplots(1,3, sharex=False, sharey=False, figsize = (20,6))
plt.subplot(1,3,1)
# CardOwnership VS Tot Population less than India and China ( pop < 1k )
sns.regplot(data=data_2014_piv_merged[data_2014_piv_merged.pop_tot < 1000], x='avg_card_own', y='pop_tot', fit_reg=True)
plt.title('Cards Ownership VS Total Population', size=20)
plt.subplot(1,3,2)
sns.regplot(data=data_2014_piv_merged, x='avg_card_own', y='urb_pop', fit_reg=True)
plt.title('Cards Ownership VS Urban Pop. Percentage', size=20)
plt.subplot(1,3,3)
sns.regplot(data=data_2014_piv_merged[data_2014_piv_merged.pop_tot < 1000], x='urb_pop', y='pop_tot', fit_reg=True)
plt.title('Total Pop. VS Urban Pop. Percentage', size=20)


plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(data_2014_piv_merged.corr())
plt.title('Dataframe Correlation Heatmap', size = 20)
plt.show()
```

We excluded China and India as outliers for Total Population in order to reduce skeweness in the scatterplot.

It is worth noting, but pretty obvious, that there is a strong correlation between Average Cards Ownership and % of Urban Population.
