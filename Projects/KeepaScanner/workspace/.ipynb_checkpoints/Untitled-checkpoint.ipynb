```python
import pandas as pd
import feedparser
import csv

```

```python


def feed_reader():
    #opening archive_lengtt.txt and saving contents to a variable
    with open("archive.csv", "r") as archive_csv:
        #importing feeds from provided link (Keepa) 
        NewsFeed = feedparser.parse("https://dyn.keepa.com/v2/user/rss/?feed=2u9m3j720a4r195m0rbahnc8e45nlab4")
        #list of new feeds
        current_entry = NewsFeed.entries
        last_feed_date = current_entry[-1]["published"]
        feed_lenght = len(current_entry)
        # length of current stored feeds read from a text CSV archive we will overwrite this at every new item
        archive = pd.read_csv(archive_csv)
        archive_lenght = len(archive)
        last_time_stamp = archive[-1][0]
        last_item_index = archive[-1][3]

        # defining index which funcion will use to detect the new feeds not stored already
        
    
        
        #if these two variables are equal, no new feeds are detected
        
        
        if  last_feed_date == last_time_stamp :
             print("NO NEWS FEED DETECTED, RESTARTING SCAN") 

            #if archive len is 0, archive is void and it need to be generated with headers too.
            
        else:
            
            if archive_lenght == 0:
         
                    with open("archive.csv", "a", newline='') as archive_csv:
                        fieldnames = ["timestamp", "title", "price", "index"]
                        writer = csv.DictWriter(archive_csv, fieldnames = fieldnames)
                        writer.writeheader()
                        starting_index = 0
                    for entry in current_entry[starting_index]:
                        
                        new_item = {}


                        new_feed_title = NewsFeed.entries[starting_index]["title"]
                        new_feed_description = NewsFeed.entries[starting_index]["description"]
                        new_feed_published = NewsFeed.entries[starting_index]["published"]
                        
                        new_item["title"] = new_feed_title
                        try :
                            money_index = new_feed_description.index("€")
                    
                        except ValueError:
                            money_index = new_feed_description.index("£") 

                        price = new_feed_description[money_index:]
                        new_item["price"] = price
                        new_item["timestamp"] = new_feed_published
                
                        print("NEW ITEM DETECTED")
                        print(str(new_feed_title))
                        print(str(new_feed_description))
                        print(str(new_feed_published))
                        print("\n")
                
                               
                        with open("archive.csv", "a", newline='') as archive_csv:
                            fieldnames = ["timestamp", "title", "price", "index"]
                            writer = csv.DictWriter(archive_csv, fieldnames = fieldnames)
                        
                            writer.writerow(new_item)
                        starting_index+=1
                             
                      #  with open("archive_length.txt", "w") as archive_lenght:
                       #     archive_lenght.write(str(current_lenght))
                        #with open("last_index.txt", "w") as last_index:
                         #   last_index.write(str(new_feed_published))
                        
                    
            elif archive_lenght > 0 :
                starting_index = int(last_item_index) + 1
                for entry in current_entry[starting_index:]:
                    print("NEW ITEM DETECTED")
                    print(NewsFeed.entries[starting_index]["title"])
                    print(NewsFeed.entries[starting_index]["description"])
                    print(NewsFeed.entries[starting_index]["published"])
                    
                    print("\n")
                
                
                    # defining dict variable to store item values
                    new_item = {}

                    #assigning values to new_item dict
                    new_feed_title = NewsFeed.entries[starting_index]["title"]
                    new_feed_description = NewsFeed.entries[starting_index]["description"]
                    new_feed_published = NewsFeed.entries[starting_index]["published"]
                    new_item["index"] = archive_lenght - 1
                    new_item["title"] = new_feed_title
                
                    try :
                        money_index = new_feed_description.index("€")
                
                    except ValueError:
                        money_index = new_feed_description.index("£") 

                    price = new_feed_description[money_index:]
                    new_item["price"] = price
                    new_item["timestamp"] = new_feed_published
                
                
                    with open("archive.csv", "a", newline='') as archive_csv:
                        fieldnames = ["index", "title", "price", "timestamp"]
                        writer = csv.DictWriter(archive_csv, fieldnames = fieldnames)
                        
                        writer.writerow(new_item)
                    starting_index+=1
```
